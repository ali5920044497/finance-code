#قیمت گذاری اختیار 

<div dir="rtl">

در این مدل برای قیمت گذاری انواع خاصی از محصولات مالی به نام "اختیارها" در نظر گرفته شده است. اینها یک نوع رایج از محصولات مالی هستند و در دسته "مشتقات مالی" قرار می گیرند. این مستندات فرض می‌کند که خواننده قبلاً با اصطلاحات اختیارهاآشنا است. این مدل‌ها عمدتاً انواعی از چارچوب گزینه بلک شولز مرتون (که در مجموع «ژانر بلک شولز» یا «شولز مشکی عمومی» نامیده می‌شوند) هستند که برای قیمت‌گذاری اختیارها اروپایی (اختیارهایی که فقط در یک مقطع زمانی قابل اعمال هستند) استفاده می‌شوند. این کتابخانه همچنین شامل تقریب‌هایی برای ارزش اختیارهای آمریکایی (اختیارهایی که می‌توانند قبل از تاریخ انقضا اعمال شوند).

فرمول های قیمت گذاری:

1. اختیارهای سهام BlackScholes() (بدون سود سهام)
2. دارایی‌های  با سود سود مستمر (اختیارهای شاخص)
3. اختیارهای کالا 
4. اختیارهای FX 
5.  اختیارهای آسیایی در مورد کالاها
6.  اختیار های گسترش (تقریبا کرک)
7. اختیارهای آمریکایی () آمریکایی
8. اختیارهای کالای آمریکایی

فرمول های نوسانات ضمنی:

9. ماشین حساب نوسانات ضمنی برای اختیارهای اروپایی
10.   محاسبه کننده نوسانات ضمنی برای اختیارهای کالاهای اروپایی
11.   ماشین حساب فرار ضمنی برای اختیارهای آمریکایی
11.   ماشین حساب نوسانات ضمنی برای اختیارهای کالاهای آمریکایی

فرمول تعمیم یافته بلک شولز را می توانید در زیر مشاهده کنید. در حالی که این فرمول ها ممکن است در نگاه اول پیچیده به نظر برسند، اکثر شرایط را می توان به عنوان بخشی از قرارداد اختیار معامله یافت یا قیمت هایی را که به راحتی در بازار در دسترس هستند. تنها اصطلاحی که محاسبه آن دشوار است، نوسانات ضمنی (σ) است. نوسانات ضمنی معمولاً با استفاده از قیمت سایر گزینه هایی که اخیراً معامله شده اند محاسبه می شود.

قیمت خرید
$$\begin{equation}
C = Fe^{(b-r)T} N(D_1) - Xe^{-rT} N(D_2)
\end{equation}$$

قیمت فروش
$$\begin{equation}
P = Xe^{-rT} N(-D_2) - Fe^{(b-r)T} N(-D_1)
\end{equation}$$

در این جا
$$\begin{equation}
D_1 = \frac{ln\frac{F}{X} + (b+\frac{V^2}{2})T}{V*\sqrt{T}}
\end{equation}$$

$$\begin{equation}
D_2 = D_1 - V\sqrt{T}
\end{equation}$$

**قیمت سهام**:
قیمت دارایی پایه در تاریخ ارزیابی. $S$ معمولاً برای نشان دادن یک قیمت لحظه ای استفاده می شود، $F$ یک قیمت آتی.


 **قیمت اعمال**:
قیمت تصویب شده اختیار

**تاریخ سررسید**:زمان انقضا بر حسب سال این را می توان با مقایسه زمان بین تاریخ انقضا و تاریخ ارزش گذاری محاسبه کرد.

**تاریخ ارزش گذاری**:
تاریخی که در آن اختیار ارزش گذاری می شود. به عنوان مثال، ممکن است تاریخ امروز باشد اگر اختیاری که امروز برای ما ارزش گذاری می شود.

**نوسان**:
این عامل را معمولا نمی توان به طور مستقیم در بازار مشاهده کرد. اغلب با نگاه کردن به قیمت‌های معاملات اخیر اختیار معامله و حل معکوس معادله سبک بلک اسکولز برای یافتن نوسانی که منجر به قیمت مشاهده‌شده می‌شود، محاسبه می‌شود. این معمولاً با حرف یونانی sigma,σ مخفف می شود، اگرچه V در اینجا برای سازگاری با کد زیر استفاده می شود.

**بازده مستمر**:
که در مدل مرتون استفاده می شود، این بازده مداوم اوراق بهادار زیربنایی است. به دارندگان اختیار معامله معمولاً تا زمانی که از اختیار معامله استفاده نکنند، سود سهام یا سایر پرداخت ها پرداخت نمی شود. در نتیجه این عامل ارزش یک گزینه را کاهش می دهد.

**نرخ بدون ریسک**:
این امید سرمایه سرمایه گذاری بدون هزینه است. این معمولاً بازده قرضه دولتی بانکی یا نرخی است که بانک های بزرگ بین خود وام گیرنده  تقریبی می شود. نرخ به مدت زمان جریان نقدی دارد. به عنوان مثال، قرضه 10 ساله بدون گزارش احتمالا دارای ارزش متفاوتی نسبت به کتاب قرضه 20 ساله بدون گزارش است.

<div dir="rtl">

```python
# T = maturity time 
# S = stock price 
# K = strike price 
# r = risk-free rate 
# sigma - volatility 
# q: - divedn yeild 
# options - call or put 
import numpy  as np
import scipy.stats as si

def black_scholes(S, K, T, r,q, sigma, option):

        d1 = (np.log(S / K) + (r - q + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
        d2 = (np.log(S / K) + (r - q - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))

        if option == 'call':
            result = (S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0))
        if option == 'put':
            result = (K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0) - S * si.norm.cdf(-d1, 0.0, 1.0))

        return round(result,4)
# T = maturity time 
T = 30/365

# S = stock price 
S = 100

# K = strike price 
K = 99

# r = risk-free rate 
r = 0.01

# q: - divedn yeild 
q = 0

# sigma - volatility 
sigma = 0.10

# options - call or put 
option = 'put'

black_scholes(S, K, T, r,q, sigma, option)
```

#درخت دو جمله ای
<div dir="rtl">
درخت دو جمله ای یک تکنیک محبوب برای قیمت گذاری اختیارها هستند. این روش بر این فرض تکیه دارد که قیمت سهام از یک راهپیمایی تصادفی پیروی می کند و در هر مرحله زمانی، احتمال معینی برای افزایش یا کاهش درصد معینی دارد. علاوه بر این، مستلزم این فرض است که هیچ فرصت آربیتراژی وجود ندارد. درخت دو جمله ای نشان دهنده مسیرهای مختلفی است که قیمت سهام می تواند دنبال کند. در حد، این مدل به مدل بلک شولز مرتون همگرا می شود که در اینجا توضیح داده شده است.

**درخت دو جمله ای یک مرحله ای**:
سهامی را در نظر بگیرید که در حال حاضر با قیمت 50 معامله می شود. ما علاقه مند به قیمت گذاری یک اختیار خرید اروپایی با اعتصاب 50 و سررسید شش ماه دیگر هستیم. می دانیم که در سررسید، سهم یا تا 55 حرکت می کند یا به 45 کاهش می یابد. اگر قیمت سهام به سمت پایین حرکت کند، اختیار دارای ارزش صفر خواهد بود.

با فرض یک دنیای خنثی ریسک، می‌توانیم پرتفویی از خود سهام و اختیار انتخابی را تنظیم کنیم تا از قبل ارزش پرتفوی را در سررسید بدانیم. بنابراین، باید دلتای سهامی را که باید نگه داریم تا پرتفوی بدون ریسک محاسبه کنیم. می توانیم دلتا را به صورت محاسبه کنیم

$$\Delta=\frac{c_u-c_d}{S_0u-S_0d},$$
که در آن $c_u$ و $c_d$ ارزش گزینه است با فرض اینکه سهام به ترتیب به سمت بالا حرکت کند و $S_0u$ و $S_0d$ ارزش سهام در صورت حرکت به سمت پایین است. از این رو، دلتا به عنوان داده می شود

$$\Delta=\frac{5-0}{55-45}=0.5.$$

فرمول زیر امکان قیمت گذاری یک اختیار را با استفاده از درخت دو جمله ای یک مرحله ای فراهم می کند. مقدار یک اختیار$o$ به صورت تعریف شده است

$$o = e^{-rT}[po_u+(1-p)o_d]$$

که در آن $o_u$ ارزش گزینه را در صورت حرکت رو به بالا سهام و $o_d$ در صورت حرکت به سمت پایین سهام با احتمال حرکت رو به بالا $p$ تعریف می‌کند.
$$p=\frac{e^{rT}-d}{u-d}$$

<div dir="rtl">

```python
import math
def exp(x):
    return math.exp(x)
S0 = 50
K = 50
S0_u = 55
S0_d = 45
r= 0.05
T = 0.5

c_u = max(S0_u-K,0)
c_d = max(S0_d-K, 0)

delta = (c_u-c_d)/(S0_u-S0_d)

pf_value_u = (S0_u*delta-c_u)*exp(-r*T)
pf_value_d = (S0_d*delta-c_d)*exp(-r*T)

c = -(pf_value_u-(S0*delta))
print(c)
```

```python
S0 = 50
K = 50
r = 0.05
T = 0.5
u = 1.1
d = 0.9
S0_u = S0*u
S0_d = S0*d

#Call
o_u = max(S0_u-K,0)
o_d = max(S0_d-K, 0)

#Put - uncomment the following lines to calculate the value of a put option instead of a call option
#o_u = max(K-S0_u,0)
#o_d = max(K-S0_d, 0)

p = (exp(r*T)-d)/(u-d)

o = exp(-r*T)*(p*o_u+(1-p)*o_d)
print(o)
```

# قیمت گذاری اختیار آسیایی
<div dir="rtl">
اختیار آسیایی نوعی اختیار است که در آن سود به میانگین قیمت دارایی پایه در یک دوره زمانی معین بستگی دارد، برخلاف اختیار های استاندارد (آمریکایی و اروپایی) که در آن بازده به قیمت دارایی پایه در یک نقطه خاص بستگی دارد. در زمان (بلوغ). این اختیار های به خریدار اجازه می‌دهد تا دارایی پایه را به‌جای قیمت لحظه‌ای، با قیمت متوسط خریداری کند (یا بفروشد).

همچنین چندین روش برای محاسبه میانگین وجود دارد. به طور دقیق تر، میانگین بخش اختیارهای آسیایی می تواند هندسی یا حسابی باشد.

قیمتاختیارهای  هندسی آسیایی آسان است زیرا فرمول های تحلیلی وجود دارد. این به سادگی به این دلیل است که میانگین هندسی یک لایه زیرین دارای توزیع لگ نرمال است. بنابراین، می‌توانید از مدل بلک شولز با نوسانات و سطوح نرخ بهره خاص استفاده کنید، همانطور که در تجارت اختیارهای  از Frans de Weert نشان داده شده است.

این واقعیت را تأیید می کند که نوسانات میانگین مشاهدات همیشه کمتر از نوسانات خود قیمت سهم است. اختیارهای  آسیایی حسابی بسیار رایج‌تر هستند و فرمول‌های بسته‌ای برای قیمت‌گذاری ندارند، زیرا میانگین حسابی متغیرهای log-normal نرمال نیست.

سپس قیمت‌گذاری اختیارهای آسیایی حسابی بر شبیه‌سازی مونت کارلو با مدل‌های لبخند مناسب مانند نوسانات محلی تکیه می‌کند. همچنین می توانید از تکنیک تطبیق لحظه ای استفاده کنید. این شامل محاسبه دقیق دو لحظه اول آسیایی و سپس تطبیق آنها می شود و این فرض را ایجاد می کند که آنها نتیجه یک گزارش طبیعی هستند. هنگامی که این گزارش خاص را عادی تشخیص دادید، در نهایت از فرمول B&S استفاده می کنید.

<div dir="rtl">

```python
#Module Imports
import numpy as np
import numpy.random as npr
import math
import scipy as sp
from scipy import stats
'''DEFINITION OF VARIABLES
    S0 - Stock Price at T=0
    E - Strike Price
    T - Time in Years
    R - Risk Free Rate
    SIGMA - Volatility
    DT - Time Step = T/N
    DF - Discount Factor = e^-RT
    I - Number of Simulations
    P - Discrete Sampling Frequency for Asian Options 
        252/Annual, 126/SemiAnnual, 63/Quarterly, 21/Monthly, 1/Continuous
'''

S0 = 100
E=100
T=1
R=0.05
SIGMA=0.20
I=10000
P= 21 #Discrete Sampling Frequency 252/Annual, 126/SemiAnnual, 63/Quarterly, 21/Monthly, 1/Continuous
N=252
'''OPTION VALUATION - W/ ANTITHETIC VARIANCE REDUCTION W/ MILSTEIN SCHEME - 
ASIAN OPTIONS - FIXED AND FLOATING STRIKE'''
def option_valuation(S0, E, T, N, SIGMA, R, I, P):    
    DT = T/N   #Time Step
    DF = math.exp(-R*T)  #Discount Factor    
#GENERATE RANDOM NUMBERS - ANTITHETIC VARIANCE REDUCTION
    PHI = npr.standard_normal((N,int(I/2))) 
    PHI = np.concatenate((PHI, -PHI), axis=1)     
#SET UP EMPTY ARRAYS AND SET INITIAL VALUES    
    S = np.zeros_like(PHI)  #Array to Capture Asset Value Path
    S[0] = S0
    P_AVG=np.zeros_like((S))  #Array to Capture Arithmetic Average Sample
    G_AVG=np.zeros_like((S))  #Array to Capture Geometric Average Sample
#CREATE FOR LOOP TO GENERATE SIMULATION PATHS - MILSTEIN METHOD
    for t in range (1,N):
        S[t]=S[t-1]*(1+R*DT+(SIGMA*PHI[t]*np.sqrt(DT))+(SIGMA**2)*(0.5*(((PHI[t]**2)-1)*DT)))
#Heaviside Function to Determine When to Take an Average
#On sample date the average is taken and stored in the appropriate array
        Mod = int(t) % P 
        if Mod == 0:
            P_AVG [t-1] = np.mean(S[(t-(P)):t], axis=0)
            G_AVG [t-1] = sp.stats.gmean(S[(t-P):t], axis=0)
            
        P_AVG[-1] = np.mean(S[(-P):N], axis=0)
        P_AVG_Payoff = np.sum(P_AVG[0:N], axis=0) / (N/P)
        
        G_AVG[-1] = sp.stats.gmean(S[(-P):N], axis=0)
        G_AVG_Payoff = np.sum(G_AVG[0:N], axis=0) / (N/P)  
 
#Calculation of Discounted Expected Payoff for Asian Options - Arithmetic Mean 
    Call_Value_Asian = DF * np.sum(np.maximum((P_AVG_Payoff) - E, 0)) / I
    print( "Value of Fixed Strike Asian Call Option - Arithmetic Average =  %.3f" %Call_Value_Asian)
    Put_Value_Asian = DF * np.sum(np.maximum(E - (P_AVG_Payoff), 0)) / I
    print( "Value of Fixed Strike Asian Put Option - Arithmetic Average = %.3f" %Put_Value_Asian) 

#Calculation of Discounted Expected Payoff for Asian Options - Geometric Mean
    Call_Value_Asian_GEO = DF * np.sum(np.maximum((G_AVG_Payoff) - E, 0)) / I
    print( "Value of Asian Fixed Strike Call Option - Geometric Average = %.3f" %Call_Value_Asian_GEO)
    Put_Value_Asian_GEO = DF * np.sum(np.maximum(E - (G_AVG_Payoff), 0)) / I
    print( "Value of Asian Fixed Strike Put Option - Geometric Average = %.3f" %Put_Value_Asian_GEO)

#Calculation of Discounted Expected Payoff for Asian Options - Geometric Mean - Floating Strike
    Call_Value_Asian_GEO_Float_Strike = DF * np.sum(np.maximum(S[-1] - (G_AVG_Payoff), 0)) / I
    print( "Value of Asian Floating Strike Call Option - Geometric Average = %.3f" %Call_Value_Asian_GEO_Float_Strike)
    Put_Value_Asian_GEO_Float_Strike = DF * np.sum(np.maximum((G_AVG_Payoff) - S[-1], 0)) / I
    print( "Value of Asian Floating Strike Put Option - Geometric Average = %.3f" %Put_Value_Asian_GEO_Float_Strike)

#Calculation of Discounted Expected Payoff for Asian Options - Arithmetic Mean - Floating Strike 
    Call_Value_Asian_Float_Strike = DF * np.sum(np.maximum(S[-1] - (P_AVG_Payoff), 0)) / I
    print( "Value of Asian Floating Strike Call Option - Arithmetic Average = %.3f" %Call_Value_Asian_Float_Strike)
    Put_Value_Asian_Float_Strike = DF * np.sum(np.maximum((P_AVG_Payoff) - S[-1], 0)) / I
    print( "Value of Asian Floating Strike Put Option - Arithmetic Average = %.3f" %Put_Value_Asian_Float_Strike) 
option_valuation(S0, E, T, N, SIGMA, R, I, P)
```

#اختیارمانع(Barrier Option)
<div dir="rtl">
اختیار مانع نوعی قرارداد اختیار مشتقه است که بازده آن به ارزش دارایی پایه بستگی دارد. به عبارت دیگر، بازپرداخت تنها زمانی اعمال می‌شود که دارایی زیربنای اختیار مانع به قیمت از پیش تعیین‌شده مشخص‌شده در قرارداد اختیار معامله برسد یا از آن فراتر رود.

در صورتی که قیمت دارایی پایه از قیمت معینی فراتر رود، یک اختیار مانع می‌تواند بی‌ارزش منقضی شود (اختیار حذفی)، بنابراین سود را برای دارنده اختیار محدود می‌کند و ضرر را برای گزینه‌نویس (فروشنده) محدود می‌کند. گزینه مانع می‌تواند یک گزینه غیرفعال باشد، به این معنی که تا زمانی که دارایی پایه به قیمت خاصی برسد، فاقد هرگونه ارزشی است.
اختیار های مانع به موارد زیر طبقه بندی می شوند:

**Knock-in barrier option**

یک اختیار مانع Knock-in یک گزینه مانع است که در آن حقوق مرتبط با رسیدن یک دارایی پایه به قیمت معینی شروع می شود. این بدان معناست که دارنده می‌تواند تنها در لحظه و پس از رسیدن قیمت به یک سطح خاص در بازار آزاد از اختیار معامله استفاده کند.

اگر هر زمانی در طول مدت قرارداد اختیار مانع به سطح قیمتی ضربه‌ای وارد شود، آنگاه به یک اختیار وانیلی تبدیل می‌شود و قیمت مناسبی دارد. اگر هرگز به سطح قیمت ضربه‌ای نرسید، اختیار مانع ضربه‌ای بی‌ارزش منقضی می‌شود.

اختیار های مانع ضربه ای بیشتر به اختیار های بالا و داخل یا پایین و در طبقه بندی می شوند. در یک اختیار مانع افزایشی، قرارداد اختیار معامله تنها زمانی شروع می شود که قیمت دارایی پایه از مانع قیمتی از پیش تعیین شده فراتر رود. برعکس، اگر یک اختیار مانع نزولی باشد، زمانی که ارزش دارایی زیربنایی به زیر قیمت مانع تعیین‌شده اولیه می‌رسد، معتبر می‌شود.

**Knock-out barrier option**

تا آنجا که به اختیار های مانع Knock-out مربوط می شود، اعتبار آنها زمانی که دارایی پایه در طول افق زمانی قرارداد به مانع برخورد کند، پایان می یابد. اختیار مانع حذفی نیز می‌توانند بیشتر به اختیار بالا و پایین یا پایین و بیرون تجزیه شوند.
<div dir="rtl">

```python
import numpy as np
import pylab as plt
import math
import scipy.stats as si
def BSM(x,t,T,k,r,sig):
    tau=T-t
    d1=(np.log(x/k) + (r+0.5*(sig*sig))*tau)/(sig*np.sqrt(tau))
    d2=d1-(sig*np.sqrt(tau))
    bsm = x*si.norm.cdf(d1,0.0,1.0) -k*np.exp(-r*tau)*si.norm.cdf(d2,0.0,1.0)
    print('BSM Price :',bsm,'\n')
def MonteCarlo_Antithetic_BarrierOption(K,n,N,sig,r,s0,V):
    Sin=np.zeros(K)
    Sout=np.zeros(K)
    Sinn=np.zeros(K)
    Soutt=np.zeros(K) 
    std = sig/np.sqrt(N) 
    mu=(r-0.5*(std*std*N))/N

    for j in range(K):
        np.random.seed(j+1)
        xi = np.random.normal(mu,std,n)
        S,DIi,DIo=[s0],0,1
        for i in range(n):
            S.append(S[-1]*np.exp(xi[i]))
            if (S[-1]<=V):DIi,DIo=1,0
        Sin[j]=np.exp(-r*n/N)*DIi*max(S[-1]-k,0)
        Sout[j]=np.exp(-r*n/N)*DIo*max(S[-1]-k,0)

        S,DIi,DIo=[s0],0,1
        for i in range(n): 
            S.append(S[-1]*np.exp(2*mu-xi[i]))
            if (S[-1]<=V):DIi,DIo=1,0
        Sinn[j]=np.exp(-r*n/N)*DIi*max(S[-1]-k,0)
        Soutt[j]=np.exp(-r*n/N)*DIo*max(S[-1]-k,0)
        

    din=np.mean(Sin)
    dout=np.mean(Sout)
    print('Simulations = ',K,'\nMonte Carlo\n',' Down and in barrier option value : ',din,'\n  Down and out barrier option value: ',dout)
    print('  Sum of both barrier option values: ',din+dout,'\n')
    din=np.mean((Sinn+Sin)/2)
    dout=np.mean((Soutt+Sout)/2)
    print('Antithetic variable \n  Down and in barrier option value : ',din,'\n  Down and out barrier option value: ',dout)
    print('  Sum of both barrier option values: ',din+dout,'\n\n')
k,r,sig,n,V,N,s0=50,5e-2,0.3,252,45,252,50
BSM(50,0,1,k,r,sig)
for i in range(2,6):MonteCarlo_Antithetic_BarrierOption(10**i,n,N,sig,r,s0,V)
```

# اختیار Loop-back
<div dir="rtl">

همچنین به عنوان یک اختیار آینده نگری شناخته می شود، یک اختیار نگاه به دارنده به دارنده این امکان را می دهد که در هنگام تعیین زمان استفاده از اختیار خود، از دانستن تاریخچه استفاده کنند. این نوع اختیار، عدم قطعیت های مرتبط با زمان ورود به بازار را کاهش می دهد و احتمال بی ارزش شدن اختیار را کاهش می دهد.
<div dir="rtl">

```python
def MonteCarlo_Loopback(K,n,N,sig,r,s0):
    Sin=np.zeros(K)
    std = sig/np.sqrt(N)
    mu=(r-0.5*(std*std*N))/N
    
    for j in range(K):
        np.random.seed(j+1)
        xi = np.random.normal(mu,std,n)
        S=[s0]
        for i in range(n):S.append(S[-1]*np.exp(xi[i]))
        Sin[j]=np.exp(-r*n/N)*max(S[-1]-np.min(S),0)
    lpbk=np.mean(Sin)
    print('Simulations :',K)
    print('Loop-back Option Value:',lpbk)
k,r,sig,n,N,s0=100,6e-2,0.25,252,252,100
for i in range(2,6):MonteCarlo_Loopback(10**i,n,N,sig,r,s0)
```
